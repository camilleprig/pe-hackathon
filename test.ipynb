## Importation des données

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('fichier.csv')
df.head()
df = df[df.default_flag == 1]
```

## Partie Lise

```python
dfl = df
dfl2 = dfl[["pl_name", "discoverymethod", "disc_year", "disc_facility"]]
```

Nous allons étudier les méthodes de découverte des planètes.

```python
plt.hist(dfl2["discoverymethod"])
```

Ce n'est pas très lisible. On remarque qu'une méthode se démarque, regardons laquelle à l'aide d'un piechart.

```python
dfl2["discoverymethod"].value_counts(normalize = True).plot(kind = "pie")
```

La méthode la plus utilisée est de loin le transit, suivi de vitesse radiale.
Étudions à présent les années de découverte.

```python
dfl2["disc_year"].value_counts(normalize = True).plot(kind = "pie")
```

On voit que les années avec le plus de découvertes sont 2016 puis 2014 puis 2021.
Passons aux installations des découvertes.

```python
dfl2["disc_facility"].value_counts(normalize = True).plot(kind = "pie")
```

On voit ici que Kepler est très majoritaire.
